```python
import pandas as pd
```

```python
base_results_df = pd.read_csv('../data/MDataFiles_Stage1/MRegularSeasonCompactResults.csv')
teams_df = pd.read_csv('../data/MDataFiles_Stage1/MTeams.csv')
```

```python
add_home_team_df = base_results_df.merge(teams_df,how = 'left',left_on='WTeamID',right_on='TeamID')
add_away_team_df = add_home_team_df.merge(teams_df,how = 'left',left_on='LTeamID',right_on='TeamID')
limit_column_names_df = add_away_team_df[['Season','DayNum','WTeamID','TeamName_x','WLoc','WScore','LTeamID','TeamName_y','LScore','NumOT']]
limit_column_names_df.columns = ['Season','DayNum','WTeamID','WTeamName','WLoc','WScore','LTeamID','LTeamName','LScore','NumOT']
clean_results_df = limit_column_names_df
```

```python
cincy_results_df = clean_results_df[(clean_results_df.WTeamName == 'Cincinnati') | (clean_results_df.LTeamName == 'Cincinnati')]
cincy_results_df[cincy_results_df.LTeamName != 'Cincinnati'].groupby(['LTeamName']).count().reset_index()[['LTeamName','WTeamID']].sort_values(by = ['WTeamID'],ascending= False).head(10)
```

```python
cincy_results_df[cincy_results_df.WTeamName != 'Cincinnati'].groupby(['WTeamName']).count().reset_index()[['WTeamName','WTeamID']].sort_values(by = ['WTeamID'],ascending= False).head(10)
```
