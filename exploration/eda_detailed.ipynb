```python
import pandas as pd
import numpy as np
```

```python
base_results_df = pd.read_csv('../data/MDataFiles_Stage1/MRegularSeasonDetailedResults.csv')
teams_df = pd.read_csv('../data/MDataFiles_Stage1/MTeams.csv')
```

```python
add_home_team_df = base_results_df.merge(teams_df,how = 'left',left_on='WTeamID',right_on='TeamID')
add_away_team_df = add_home_team_df.merge(teams_df,how = 'left',left_on='LTeamID',right_on='TeamID')
```

```python
add_away_team_df
```

```python
add_away_team_df.columns
```

```python
limit_column_names_df = add_away_team_df[['Season','DayNum','WTeamID','TeamName_x','WLoc','WScore','LTeamID','TeamName_y','LScore','NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF']]
limit_column_names_df.columns = ['Season','DayNum','WTeamID','WTeamName','WLoc','WScore','LTeamID','LTeamName','LScore','NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF']
clean_results_df = limit_column_names_df
```

```python
results_by_team_win_df = clean_results_df.copy()
results_by_team_win_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT', 'FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR',
       'Ast', 'TO', 'Stl', 'Blk', 'PF', 'OppFGM', 'OppFGA', 'OppFGM3', 'OppFGA3','OppFTM', 'OppFTA', 'OppOR', 'OppDR', 'OppAst', 'OppTO', 'OppStl', 'OppBlk', 'OppPF']
results_by_team_win_df['Result'] = 'W'
```

```python
def new_loc(row):
    new_loc = ''
    if row['Loc'] == 'H':
        new_loc = 'A'
    elif row['Loc'] == 'A':
        new_loc = 'H'
    else :
        new_loc = 'N'
    return new_loc

def win(row):
    win = 0
    if row['Result'] == 'W':
        win = 1
    elif row['Result'] == 'L':
        win = 0
    else :
        win = 0
    return win
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF' ,'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF']]
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF' ,'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF']]
results_by_team_loss_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT', 'FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR',
       'Ast', 'TO', 'Stl', 'Blk', 'PF', 'OppFGM', 'OppFGA', 'OppFGM3', 'OppFGA3','OppFTM', 'OppFTA', 'OppOR', 'OppDR', 'OppAst', 'OppTO', 'OppStl', 'OppBlk', 'OppPF']
results_by_team_loss_df['Result'] = 'L'
results_by_team_loss_df['Loc'] = results_by_team_loss_df.apply(new_loc, axis=1)
```

```python
results_by_team_df = pd.concat([results_by_team_win_df,results_by_team_loss_df])
```

```python
results_by_team_df.columns
```

```python
results_by_team_df['Win'] = results_by_team_df.apply(win, axis=1)
```

```python
results_by_team_df['FGPerc'] = results_by_team_df['FGM'] / results_by_team_df['FGA']
results_by_team_df['3PtFGPerc'] = results_by_team_df['FGM3'] / results_by_team_df['FGA3']
results_by_team_df['OppFGPerc'] = results_by_team_df['OppFGM'] / results_by_team_df['OppFGA']
results_by_team_df['Opp3PtFGPerc'] = results_by_team_df['OppFGM3'] / results_by_team_df['OppFGA3']
results_by_team_df['AsstTORatio'] = results_by_team_df['Ast'] / results_by_team_df['TO']
results_by_team_df['OppAsstTORatio'] = results_by_team_df['OppAst'] / results_by_team_df['OppTO']
results_by_team_df['RebRatio'] = (results_by_team_df['OR'] + results_by_team_df['DR']) / (results_by_team_df['OppOR'] + results_by_team_df['OppDR'])
results_by_team_df['ScoreDiff'] = results_by_team_df['Score'] - results_by_team_df['OppScore']
results_by_team_df['FTPerc'] = results_by_team_df['FTM'] / results_by_team_df['FTA']
results_by_team_df['OppFTPerc'] = results_by_team_df['OppFTM'] / results_by_team_df['OppFTA']
results_by_team_df['TotalReb'] = results_by_team_df['OR'] + results_by_team_df['DR']
results_by_team_df['OppTotalReb'] = results_by_team_df['OppOR'] + results_by_team_df['OppDR']
results_by_team_df['PFRatio'] = results_by_team_df['PF'] / results_by_team_df['OppPF']
results_by_team_df['FTRatio'] = results_by_team_df['FTA'] / results_by_team_df['OppFTA']
results_by_team_df['Game'] = 1

```

```python
team_season_stats_agg_df = results_by_team_df.groupby(['Season','TeamName']).agg(
    mean_Points=('Score', np.mean),
    std_Points=('Score', np.std),
    median_Points=('Score', np.median),
    mean_FGM=('FGM',np.mean),
    std_FGM=('FGM',np.std),
    median_FGM=('FGM', np.median),
    mean_FGPerc=('FGPerc',np.mean),
    std_FGPerc=('FGPerc',np.std),
    median_FGPerc=('FGPerc',np.median),
    mean_3PtFGPerc=('3PtFGPerc',np.mean),
    std_3PtFGPerc=('3PtFGPerc',np.std),
    median_3PtFGPerc=('3PtFGPerc',np.median),
    mean_OppFGPerc=('OppFGPerc',np.mean),
    std_OppFGPerc=('OppFGPerc',np.std),
    median_OppFGPerc=('OppFGPerc',np.median),        
    mean_Opp3PtFGPerc=('Opp3PtFGPerc',np.mean),
    std_Opp3PtFGPerc=('Opp3PtFGPerc',np.std),
    median_Opp3PtFGPerc=('Opp3PtFGPerc',np.median),
    mean_AsstToRatio=('AsstTORatio',np.mean),
    std_AsstToRatio=('AsstTORatio',np.std),
    median_AsstToRatio=('AsstTORatio',np.median),
    mean_OppAsstToRatio=('OppAsstTORatio',np.mean),
    std_OppAsstToRatio=('OppAsstTORatio',np.std),
    median_OppAsstToRatio=('OppAsstTORatio',np.median),  
    mean_RebRatio=('RebRatio',np.mean),
    std_RebRatio=('RebRatio',np.std),
    median_RebRatio=('RebRatio',np.median),  
    mean_ScoreDiff=('ScoreDiff',np.mean),
    std_ScoreDiff=('ScoreDiff',np.std),
    median_ScoreDiff=('ScoreDiff',np.median), 
    mean_FTPerc=('FTPerc',np.mean),
    std_FTPerc=('FTPerc',np.std),
    median_FTPerc=('FTPerc',np.median),
    mean_OppFTPerc=('OppFTPerc',np.mean),
    std_OppFTPerc=('OppFTPerc',np.std),
    median_OppFTPerc=('OppFTPerc',np.median),
    mean_OR=('OR',np.mean),
    std_OR=('OR',np.std),
    median_OR=('OR',np.median),
    mean_DR=('DR',np.mean),
    std_DR=('DR',np.std),
    median_DR=('DR',np.median),
    mean_TotalReb=('TotalReb',np.mean),
    std_TotalReb=('TotalReb',np.std),
    median_TotalReb=('TotalReb',np.median),
    mean_OppOR=('OppOR',np.mean),
    std_OppOR=('OppOR',np.std),
    median_OppOR=('OppOR',np.median),
    mean_OppDR=('OppDR',np.mean),
    std_OppDR=('OppDR',np.std),
    median_OppDR=('OppDR',np.median),    
    mean_OppTotalReb=('OppTotalReb',np.mean),
    std_OppTotalReb=('OppTotalReb',np.std),
    median_OppTotalReb=('OppTotalReb',np.median),
    mean_PF=('PF',np.mean),
    std_PF=('PF',np.std),
    median_PF=('PF',np.median),
    mean_OppPF=('OppPF',np.mean),
    std_OppPF=('OppPF',np.std),
    median_OppPF=('OppPF',np.median),
    mean_PFRatio=('PFRatio',np.mean),
    std_PFRatio=('PFRatio',np.std),
    median_PFRatio=('PFRatio',np.median),
    mean_Stl=('Stl',np.mean),
    std_Stl=('Stl',np.std),
    median_Stl=('Stl',np.median),
    mean_Blk=('Blk',np.mean),
    std_Blk=('Blk',np.std),
    median_Blk=('Blk',np.median),
    mean_OppStl=('OppStl',np.mean),
    std_OppStl=('OppStl',np.std),
    median_OppStl=('OppStl',np.median),
    mean_OppBlk=('OppBlk',np.mean),
    std_OppBlk=('OppBlk',np.std),
    median_OppBlk=('OppBlk',np.median),
    mean_TO=('TO',np.mean),
    std_TO=('TO',np.std),
    median_TO=('TO',np.median),
    mean_OppTO=('OppTO',np.mean),
    std_OppTO=('OppTO',np.std),
    median_OppTO=('OppTO',np.median),
    mean_Ast=('Ast',np.mean),
    std_Ast=('Ast',np.std),
    median_Ast=('Ast',np.median),
    mean_OppAst=('OppAst',np.mean),
    std_OppAst=('OppAst',np.std),
    median_OppAst=('OppAst',np.median),
    mean_FTA=('FTA',np.mean),
    std_FTA=('FTA',np.std),
    median_FTA=('FTA',np.median),
    mean_OppFTA=('OppFTA',np.mean),
    std_OppFTA=('OppFTA',np.std),
    median_OppFTA=('OppFTA',np.median),
    mean_FTRatio=('FTRatio',np.mean),
    std_FTRatio=('FTRatio',np.std),
    median_FTRatio=('FTRatio',np.median),
    sum_Wins=('Win',np.sum),
    sum_Games=('Game',np.sum),     
    )
```

```python
team_season_stats_agg_df = team_season_stats_agg_df.reset_index()
team_season_stats_agg_df['WinPerc'] = team_season_stats_agg_df['sum_Wins'] / team_season_stats_agg_df['sum_Games']
team_season_stats_agg_df
```
