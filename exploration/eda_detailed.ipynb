```python
import pandas as pd
import numpy as np
```

```python
base_results_df = pd.read_csv('../data/MDataFiles_Stage1/MRegularSeasonDetailedResults.csv')
teams_df = pd.read_csv('../data/MDataFiles_Stage1/MTeams.csv')
```

```python
add_home_team_df = base_results_df.merge(teams_df,how = 'left',left_on='WTeamID',right_on='TeamID')
add_away_team_df = add_home_team_df.merge(teams_df,how = 'left',left_on='LTeamID',right_on='TeamID')
```

```python
add_away_team_df
```

```python
add_away_team_df.columns
```

```python
limit_column_names_df = add_away_team_df[['Season','DayNum','WTeamID','TeamName_x','WLoc','WScore','LTeamID','TeamName_y','LScore','NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF']]
limit_column_names_df.columns = ['Season','DayNum','WTeamID','WTeamName','WLoc','WScore','LTeamID','LTeamName','LScore','NumOT', 'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF']
clean_results_df = limit_column_names_df
```

```python
results_by_team_win_df = clean_results_df.copy()
results_by_team_win_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT', 'FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR',
       'Ast', 'TO', 'Stl', 'Blk', 'PF', 'OppFGM', 'OppFGA', 'OppFGM3', 'OppFGA3','OppFTM', 'OppFTA', 'OppOR', 'OppDR', 'OppAst', 'OppTO', 'OppStl', 'OppBlk', 'OppPF']
results_by_team_win_df['Result'] = 'W'
```

```python
def new_loc(row):
    new_loc = ''
    if row['Loc'] == 'H':
        new_loc = 'A'
    elif row['Loc'] == 'A':
        new_loc = 'H'
    else :
        new_loc = 'N'
    return new_loc
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF' ,'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF']]
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT', 'LFGM', 'LFGA', 'LFGM3', 'LFGA3','LFTM', 'LFTA', 'LOR', 'LDR', 'LAst', 'LTO', 'LStl', 'LBlk', 'LPF' ,'WFGM', 'WFGA', 'WFGM3', 'WFGA3', 'WFTM', 'WFTA', 'WOR', 'WDR',
       'WAst', 'WTO', 'WStl', 'WBlk', 'WPF']]
results_by_team_loss_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT', 'FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR',
       'Ast', 'TO', 'Stl', 'Blk', 'PF', 'OppFGM', 'OppFGA', 'OppFGM3', 'OppFGA3','OppFTM', 'OppFTA', 'OppOR', 'OppDR', 'OppAst', 'OppTO', 'OppStl', 'OppBlk', 'OppPF']
results_by_team_loss_df['Result'] = 'L'
results_by_team_loss_df['Loc'] = results_by_team_loss_df.apply(new_loc, axis=1)
```

```python
results_by_team_df = pd.concat([results_by_team_win_df,results_by_team_loss_df])
```

```python
results_by_team_df.columns
```

```python
results_by_team_df['FGPerc'] = results_by_team_df['FGM'] / results_by_team_df['FGA']
results_by_team_df['3PtFGPerc'] = results_by_team_df['FGM3'] / results_by_team_df['FGA3']
results_by_team_df['OppFGPerc'] = results_by_team_df['OppFGM'] / results_by_team_df['OppFGA']
results_by_team_df['Opp3PtFGPerc'] = results_by_team_df['OppFGM3'] / results_by_team_df['OppFGA3']
results_by_team_df['AsstTORatio'] = results_by_team_df['Ast'] / results_by_team_df['TO']
results_by_team_df['OppAsstTORatio'] = results_by_team_df['OppAst'] / results_by_team_df['OppTO']
results_by_team_df['RebRatio'] = (results_by_team_df['OR'] + results_by_team_df['DR']) / (results_by_team_df['OppOR'] + results_by_team_df['OppDR'])
results_by_team_df['ScoreDiff'] = results_by_team_df['Score'] - results_by_team_df['OppScore']
```

```python
results_by_team_df.groupby(['Season','TeamName']).agg(
    mean_Points=('Score', np.mean),
    std_Points=('Score', np.std),
    median_Points=('Score', np.median),
    mean_FGM=('FGM',np.mean),
    std_FGM=('FGM',np.std),
    median_FGM=('FGM', np.median),
    mean_FGPerc=('FGPerc',np.mean),
    std_FGPerc=('FGPerc',np.std),
    median_FGPerc=('FGPerc',np.median),
    mean_3PtFGPerc=('3PtFGPerc',np.mean),
    std_3PtFGPerc=('3PtFGPerc',np.std),
    median_3PtFGPerc=('3PtFGPerc',np.median),
    mean_OppFGPerc=('OppFGPerc',np.mean),
    std_OppFGPerc=('OppFGPerc',np.std),
    median_OppFGPerc=('OppFGPerc',np.median),        
    mean_Opp3PtFGPerc=('Opp3PtFGPerc',np.mean),
    std_Opp3PtFGPerc=('Opp3PtFGPerc',np.std),
    median_Opp3PtFGPerc=('Opp3PtFGPerc',np.median),
    mean_AsstToRatio=('AsstTORatio',np.mean),
    std_AsstToRatio=('AsstTORatio',np.std),
    median_AsstToRatio=('AsstTORatio',np.median),
    mean_OppAsstToRatio=('AsstTORatio',np.mean),
    std_OppAsstToRatio=('AsstTORatio',np.std),
    median_OppAsstToRatio=('AsstTORatio',np.median),         
    )
```
