```python
import pandas as pd
import numpy as np
```

```python
base_results_df = pd.read_csv('../data/MDataFiles_Stage1/MRegularSeasonCompactResults.csv')
teams_df = pd.read_csv('../data/MDataFiles_Stage1/MTeams.csv')
```

```python
add_home_team_df = base_results_df.merge(teams_df,how = 'left',left_on='WTeamID',right_on='TeamID')
add_away_team_df = add_home_team_df.merge(teams_df,how = 'left',left_on='LTeamID',right_on='TeamID')
limit_column_names_df = add_away_team_df[['Season','DayNum','WTeamID','TeamName_x','WLoc','WScore','LTeamID','TeamName_y','LScore','NumOT']]
limit_column_names_df.columns = ['Season','DayNum','WTeamID','WTeamName','WLoc','WScore','LTeamID','LTeamName','LScore','NumOT']
clean_results_df = limit_column_names_df
```

```python
results_by_team_win_df = clean_results_df.copy()
results_by_team_win_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT']
results_by_team_win_df['Result'] = 'W'
```

```python
def new_loc(row):
    new_loc = ''
    if row['Loc'] == 'H':
        new_loc = 'A'
    elif row['Loc'] == 'A':
        new_loc = 'H'
    else :
        new_loc = 'N'
    return new_loc
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT']]
```

```python
results_by_team_loss_df = clean_results_df.copy()
results_by_team_loss_df = results_by_team_loss_df[['Season','DayNum','LTeamID','LTeamName','WLoc','LScore','WTeamID','WTeamName','WScore','NumOT']]
results_by_team_loss_df.columns = ['Season','DayNum','TeamID','TeamName','Loc','Score','OppTeamID','OppTeamName','OppScore','NumOT']
results_by_team_loss_df['Result'] = 'L'
results_by_team_loss_df['Loc'] = results_by_team_loss_df.apply(new_loc, axis=1)
```

```python
results_by_team_df = pd.concat([results_by_team_win_df,results_by_team_loss_df])
```
